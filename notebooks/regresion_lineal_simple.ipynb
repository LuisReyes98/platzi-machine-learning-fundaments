```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

Para todo modelo primero tienes que entrenar y luego ponerlo a prueba

```python
dataset = pd.read_csv('./salarios.csv')
dataset.head()
```

```python
dataset.shape
```

```python
# columna X y Y
x = dataset.iloc[:, :-1]
y = dataset.iloc[:,1]
```

```python
x.head()
```

```python
y.head()
```

```python
# dividiendo la informacion en entrenamiento y test
# test_size nos dice de que tamaño sera mi dataset de pruebas
# lo recomendable es que el tamaño de los datos de entramiento
# sea un 70% de los datos totales y los de test el 30% restante

# random_state = 0 hara que siempre tengamos el mismo grupo de datos
# 80% para entrenamiento, 20% de pruebas

X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
X_train.head()
```

```python
regressor = LinearRegression()
# se pasan los valores de entrenamiento al modelo
regressor.fit(X_train, Y_train)
```

```python
#datos de entramiento
viz_train = plt
viz_train.scatter(X_train, Y_train, color = 'blue')
viz_train.plot(X_train, regressor.predict(X_train), color = 'black')
viz_train.title('Salario vs Experiencia')

viz_train.xlabel('Experiencia')
viz_train.ylabel('Salario')
viz_train.show()
```

```python
# datos de test
viz_train = plt
viz_train.scatter(X_test, Y_test, color='red')
viz_train.plot(X_train, regressor.predict(X_train), color='black')
# viz_train.plot(X_train, regressor.predict(X_train), color='blue')
viz_train.title('Salario vs Experiencia')

viz_train.xlabel('Experiencia')
viz_train.ylabel('Salario')
viz_train.show()
```

```python
viz_train = plt
viz_train.scatter(X_test, Y_test, color='red')
viz_train.plot(X_test, regressor.predict(X_test), color='black')
# viz_train.plot(X_train, regressor.predict(X_train), color='blue')
viz_train.title('Salario vs Experiencia')

viz_train.xlabel('Experiencia')
viz_train.ylabel('Salario')
viz_train.show()
```

```python
# Score del modelo
regressor.score(X_test, Y_test)
```
